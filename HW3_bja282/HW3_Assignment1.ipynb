```python
##Assignment is incomplete, unfortunately. I struggled with the Statistics content and was unable to finish on time.
__author__= 'Ben Alpert'
from __future__ import print_function
import pandas as pd
import pylab as pl
import numpy as np
import matplotlib.pyplot as plt
%pylab inline

np.random.seed(203)
#distributions = ['pois', 'gaus', 'chisq', 'cauchy', 'lnorm', 'binomial']
```

```python
#SamplesDict = {"Chisquare" : {"desc" : "Chi-squared"}, "Poisson" : {"desc" : "Poisson"},
#         "Binomial" : {"desc" : "Binomial"}, "Normal" : {"desc" : "Normal"}}

mySize = (2000 / (np.array(range(1, 101)))).astype(int)

#Generate the samples for each of the distribution types with mean = 100
myMean = 100

df= myMean

#Create five lists to store each of the distributions sample means
chisqMeans = []
PoissonMeans = []
NormalMeans = []
BinomialMeans = []

#Used work group partners Scott Smith & Sebastian Gutierrez's notebooks for reference.
#https://github.com/sbg389/PUI2016_sbg389/blob/master/HW3_sbg389/HW3_1_sbg389.ipynb
#https://github.com/sighthnd/PUI2016_ss4977/blob/master/HW3_ss4977/Assignment1.ipynb
#Populate dictionary
for x in range(100):
    sample = np.random.chisquare(df,size=mySize[x]) #from Federica's notebook
    chisqMeans.append(sample.mean())
    sample = np.random.normal(df, 4, size=mySize[x]) #https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.normal.html
    NormalMeans.append(sample.mean())
    sample = np.random.poisson(lam=1.0, size=mySize[x]) #http://docs.scipy.org/doc/numpy/reference/generated/numpy.random.poisson.html
    PoissonMeans.append(sample.mean())
    sample = np.random.binomial(200, .5, mySize[x])#http://docs.scipy.org/doc/numpy/reference/generated/numpy.random.binomial.html
    BinomialMeans.append(sample.mean())
```

Chi-Squared

```python
#Subplot I: Mean value by sample size
pl.figure(figsize=(10, 4))
pl.subplot(121)
pl.title('Mean value by sample size')
pl.scatter(mySize, chisqMeans)
pl.ylabel('sample mean', fontsize = 10)
pl.xlabel('N', fontsize = 10)
```

Normal

```python
pl.figure(figsize=(10, 4))
pl.subplot(121)
pl.title('Mean value by sample size')
pl.scatter(mySize, NormalMeans)
pl.ylabel('sample mean', fontsize = 10)
pl.xlabel('N', fontsize = 10)
```

Poisson

```python
pl.figure(figsize=(10, 4))
pl.subplot(121)
pl.title('Mean value by sample size')
pl.scatter(mySize, PoissonMeans)
pl.ylabel('sample mean', fontsize = 10)
pl.xlabel('N', fontsize = 10)
```

Binomial

```python
pl.figure(figsize=(10, 4))
pl.subplot(121)
pl.title('Mean value by sample size')
pl.scatter(mySize, BinomialMeans)
pl.ylabel('sample mean', fontsize = 10)
pl.xlabel('N', fontsize = 10)
```

```python
#Multi-bar histogram demo: http://matplotlib.org/examples/statistics/histogram_demo_multihist.html

pl.figure(figsize(10,4))
pl.hist((chisqMeans, PoissonMeans, NormalMeans, BinomialMeans), bins=20, histtype='bar')
```
