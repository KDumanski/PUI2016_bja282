##GENERATE 100 samples of different sizes N (N>10 & N<2000) from each of 5 different distributions (500 samples in total), ## all with the same population mean. Include a Normal, a Poisson, a Binomial, a Chi-Squared distribution, and 1 more of your choice.
    ## For each sample, plot the sample mean (dependent var.) against the sample size N (independent var.) (if you want you can do it with the sample standard deviation as well).
    ## Describe the behavior you see in the plots in terms of the law of large numbers.
    ## PLOT the distributions of all sample means (together for all distributions). Mandatory: as a histogram. Optional: in any other way you think is convincing ## Extra Credit: FIT a gaussian to the distribution of means


```python
__author__= 'Ben Alpert'
from __future__ import print_function
import pandas as pd
import pylab as pl
import numpy as np
import matplotlib.pyplot as pyplot
%pylab inline

np.random.seed(203)
#distributions = ['pois', 'gaus', 'chisq', 'cauchy', 'lnorm', 'binomial']
```

```python
mySize = (2000 / (np.array(range(1, 101)))).astype(int)

#Generate the samples for each of the distribution types with mean = 100
myMean = 100
df= myMean
#Create five lists to store each of the distributions sample means
chisqMeans = []
PoissonMeans = []
NormalMeans = []
BinomialMeans = []
CauchyMeans = []

#Used work group partners Scott Smith & Sebastian Gutierrez's notebooks for reference.
#https://github.com/sbg389/PUI2016_sbg389/blob/master/HW3_sbg389/HW3_1_sbg389.ipynb
#https://github.com/sighthnd/PUI2016_ss4977/blob/master/HW3_ss4977/Assignment1.ipynb
#Populate dictionary
for x in range(100):
    sample = np.random.chisquare(df,size=mySize[x]) #from Federica's notebook
    chisqMeans.append(sample.mean())
    sample = np.random.normal(df, 4, size=mySize[x]) #https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.normal.html
    NormalMeans.append(sample.mean())
    sample = np.random.poisson(myMean, size=mySize[x]) #http://docs.scipy.org/doc/numpy/reference/generated/numpy.random.poisson.html
    PoissonMeans.append(sample.mean())
    sample = np.random.binomial(200, .5, mySize[x])#http://docs.scipy.org/doc/numpy/reference/generated/numpy.random.binomial.html
    BinomialMeans.append(sample.mean())
    sample = np.random.standard_cauchy(mySize[x])
    CauchyMeans.append(sample.mean())
        
def plots(size, means):
    #Scatter: 
    pl.figure(figsize=(10, 4))
    pl.subplot(121)
    pl.title('Mean Values by Sample Size')
    pl.scatter(size, means)
    pl.ylabel('Sample Mean', fontsize = 10)
    pl.xlabel('N', fontsize = 10)
    #Histogram:
    pl.subplot(122)
    pl.hist(means, bins = 30)
    pl.ylabel('Sample Mean', fontsize = 10)
    pl.xlabel('N', fontsize = 10)
    pl.title('Sample Mean Frequency')
    pl.show()
```

Chi-Squared

```python
plots(mySize, chisqMeans)
```

Normal

```python
plots(mySize, NormalMeans)
```

Poisson

```python
plots(mySize, PoissonMeans)
```

Binomial

```python
plots(mySize, BinomialMeans)
```

Cauchy

```python
plots(mySize, CauchyMeans)
```

```python
labels=["Chisquare", "Normal", "Poisson", "Binomial", "Cauchy"]
pl.hist((chisqMeans, NormalMeans, PoissonMeans, BinomialMeans, CauchyMeans), bins=20)
pl.legend(labels, loc=9)
```
